### Cleaning Data

Pada proses Cleaning Data ini data yang sudah dicek saat proses assessing akan dibersihkan, dimana proses pembersihan diantaranya adalah
1. Penanganan nilai NULL
2. Penghapusan nilai duplikat
3. Penanganan nilai yang tidak valid

Catatan :
Dikarenakan pada proses assessing diketahui bahwa keseluruhan data tidak ada yang duplikat, maka proses yang dilakukan hanyalah langkah 1 dan 3 saja

Dikarenakan hampir keseluruhan data pada dataset memiliki type numerik (float, int) maka pada proses penanganan nilai NULL pada data dapat menggunakan metode interpolation atau imputation dengan nilai modus, mean, median
Namun apabila nilai NULL tersebut berada pada coloumn data yang memiliki type selain numerik, maka proses penanganan nilai NULL dapat menggunakan imputation dengan nilai modus

Cleaning Data Dengan Interpolation Dan Imputation Untuk Pembersihan Data Numerik

```python
aotizhongxin = aotizhongxin.interpolate(method='linear')
dingling = dingling.interpolate(method='linear')
dongsi  = dongsi.interpolate(method='linear')
guanyuan  = guanyuan.interpolate(method='linear')
gucheng  = gucheng.interpolate(method='linear')
huairou  = huairou.interpolate(method='linear')
nongzhanguan  = nongzhanguan.interpolate(method='linear')
shunyi  = shunyi.interpolate(method='linear')
tiantan  = tiantan.interpolate(method='linear')
wanliu  = wanliu.interpolate(method='linear')
wanshouxigong  = wanshouxigong.interpolate(method='linear')
```

```python
aotizhongxin.NO2.fillna(value=aotizhongxin.NO2.mean(), inplace=True)
dingling.NO2.fillna(value=dingling.NO2.mean(), inplace=True)
dongsi.NO2.fillna(value=dongsi.NO2.mean(), inplace=True)
guanyuan.NO2.fillna(value=guanyuan.NO2.mean(), inplace=True)
gucheng.NO2.fillna(value=gucheng.NO2.mean(), inplace=True)
huairou.NO2.fillna(value=huairou.NO2.mean(), inplace=True)
nongzhanguan.NO2.fillna(value=nongzhanguan.NO2.mean(), inplace=True)
shunyi.NO2.fillna(value=shunyi.NO2.mean(), inplace=True)
tiantan.NO2.fillna(value=tiantan.NO2.mean(), inplace=True)
wanliu.NO2.fillna(value=wanliu.NO2.mean(), inplace=True)
wanshouxigong.NO2.fillna(value=wanshouxigong.NO2.mean(), inplace=True)
```

Cleaning Data Dengan Imputation Untuk Pembersihan Data Object

```python
aotizhongxin.wd.fillna(value="NE", inplace=True)
dingling.wd.fillna(value="NNW", inplace=True)
dongsi.wd.fillna(value="ENE", inplace=True)
guanyuan.wd.fillna(value="NE", inplace=True)
gucheng.wd.fillna(value="N", inplace=True)
huairou.wd.fillna(value="NW", inplace=True)
nongzhanguan.wd.fillna(value="ENE", inplace=True)
shunyi.wd.fillna(value="NNE", inplace=True)
tiantan.wd.fillna(value="ENE", inplace=True)
wanliu.wd.fillna(value="NE", inplace=True)
wanshouxigong.wd.fillna(value="NE", inplace=True)
```

## Visualization & Explanatory Analysis

Pertanyaan 1: Di lokasi manakah rata-rata tingkat polutan PM2.5 nya paling tinggi dan paling rendah?

```python
pm25 = full_data.groupby(by='station').agg({"PM2.5":'mean'}).sort_values(by='PM2.5', ascending=False).reset_index()
colors = ["#408ABF", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#408ABF"]

sns.barplot(x='PM2.5', y='station', data=pm25, palette=colors)
plt.title("Rata-rata tingkat polutan PM2.5", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

Pertanyaan 2: Bagaimana Korelasi antara Polutan NO2, CO Dan PM10 Dengan Polutan lainnya?

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(24,12))

sns.scatterplot(x='NO2', y='PM2.5', data=full_data, palette=colors, ax=ax[0][0], s=10)
ax[0][0].set_title("Polutan PM2.5", loc="center", fontsize=15)

sns.scatterplot(x='NO2', y='PM10', data=full_data, palette=colors, ax=ax[0][1], s=10)
ax[0][1].set_title("Polutan PM10", loc="center", fontsize=15)

sns.scatterplot(x='NO2', y='SO2', data=full_data, palette=colors, ax=ax[0][2], s=10)
ax[0][2].set_title("Polutan SO2", loc="center", fontsize=15)

sns.scatterplot(x='NO2', y='CO', data=full_data, palette=colors, ax=ax[1][0], s=10)
ax[1][0].set_title("Polutan CO", loc="center", fontsize=15)

sns.scatterplot(x='NO2', y='O3', data=full_data, palette=colors, ax=ax[1][1], s=10)
ax[1][1].set_title("Polutan O3", loc="center", fontsize=15)

plt.suptitle("Korelasi Polutan NO2 Dengan Polutan Lain", fontsize=30)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(24,12))

sns.scatterplot(x='CO', y='PM2.5', data=full_data, palette=colors, ax=ax[0][0], s=10)
ax[0][0].set_title("Polutan PM2.5", loc="center", fontsize=15)

sns.scatterplot(x='CO', y='PM10', data=full_data, palette=colors, ax=ax[0][1], s=10)
ax[0][1].set_title("Polutan PM10", loc="center", fontsize=15)

sns.scatterplot(x='CO', y='SO2', data=full_data, palette=colors, ax=ax[0][2], s=10)
ax[0][2].set_title("Polutan SO2", loc="center", fontsize=15)

sns.scatterplot(x='CO', y='NO2', data=full_data, palette=colors, ax=ax[1][0], s=10)
ax[1][0].set_title("Polutan NO2", loc="center", fontsize=15)

sns.scatterplot(x='CO', y='O3', data=full_data, palette=colors, ax=ax[1][1], s=10)
ax[1][1].set_title("Polutan O3", loc="center", fontsize=15)

plt.suptitle("Korelasi Polutan CO Dengan Polutan Lain", fontsize=30)
```
